```python
import sys
sys.path.append('../')

import numpy as np
import pandas as pd
import os

from picket.prepare.dataPrepare import dataPrepareTrain
from picket.filter.filtersTrain import evaluateTrainTime, printResTrainTime
```

# Generate Synthetic Dataset

```python
attr_num = 10
sample_num = 1000

for rank in [2, 4, 6, 8, 10]:
    path = '../data/synthetic_r%d/' % rank
    if not os.path.exists(path):
        os.makedirs(path)
    z_mean = 2*np.random.rand(rank)-1
    z_cov = np.eye(len(z_mean))
    z = np.random.multivariate_normal(z_mean, z_cov, sample_num)
    A = np.random.rand(attr_num, rank)
    x = np.matmul(A, z.T).T  
    y = np.random.randint(0, 2, (sample_num,))
    pd.DataFrame(x).to_csv(path+'features.csv', index=False)
    pd.DataFrame(y).to_csv(path+'class.csv', index=False)
```

# Run Train Time Outlier Detection

```python
for rank in [2, 4, 6, 8, 10]:
    dataset = 'synthetic_r%d' % rank
    dataPrepareTrain(dataset, eps_col=0.2, random_noise_level=1)
    evaluateTrainTime(dataset, ntype='random', eps_col=0.2)
    #evaluateTrainTime(dataset, ntype='system', eps_col=0.2)
```
