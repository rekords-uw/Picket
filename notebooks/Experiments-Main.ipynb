```python
import sys
sys.path.append('../')

from picket.prepare.dataPrepare import dataPrepareTrain, dataPrepareTest
from picket.filter.filtersTrain import evaluateTrainTime, printResTrainTime
from picket.filter.filtersTest import PK_model_prepare
from picket.filter.filtersTest import evaluateTestTime, printResTestTime 
```

```python
# dataset includes 'wine', 'adult', 'restaurant', 'marketing', 'titanic', 'HTRU2'
dataset = 'titanic'
# We only consider adversarial noise for purely numerical dataset
runAdv = False
# low, medium or high
noise_level = 'medium'
```

## Train-test Split and Noise Injection

```python
dataPrepareTrain(dataset, split=True)
dataPrepareTest(dataset, adv=runAdv)
```

## Outlier Detection at Training Time

```python
evaluateTrainTime(dataset, ntype='random', level=noise_level)
evaluateTrainTime(dataset, ntype='system', level=noise_level)
if runAdv:
    evaluateTrainTime(dataset, ntype='poison', modelname='lr')
    evaluateTrainTime(dataset, ntype='poison', modelname='svm')
    evaluateTrainTime(dataset, ntype='poison', modelname='nn')
```

```python
printResTrainTime(dataset, level=noise_level)
```

## Victim Sample Detection at Test Time

```python
# Train PicketNet on clean data
PK_model_prepare(dataset)
evaluateTestTime(dataset, runAdv=runAdv, level=noise_level, mix_artificial=True)
```

```python
printResTestTime(dataset, level=noise_level)
```
